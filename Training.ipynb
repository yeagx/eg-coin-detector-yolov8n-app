```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # Nano model
```

```python
!unzip EG-Coins-Detection.v3i.yolov8.zip
```

```python
!cat data.yaml
```

```python
results = model.train(
    data="data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    patience=10,
    name="eg_coins_yolov8n",
    device=0  # GPU
)
```

```python
metrics = model.val(
    data="data.yaml",
    split="test"
)

print(metrics)
```
